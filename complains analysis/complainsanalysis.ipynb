```python
import pandas as pd
import matplotlib.pyplot as plt

#complaints data
data = [
    {"category": "#webdev, #business", "reason": "This is a scam"},
    {"category": "#anime, #testing", "reason": "This user is fake"},
    {"category": "#testing", "reason": "Not trustworthy"},
    {"category": "#webdev, #business", "reason": "this is fake"},
    {"category": "#health, #gym", "reason": "this is fake"},
]

# Split entries with multiple categories into individual entries
split_data = []
for entry in data:
    categories = entry["category"].split(", ")
    for category in categories:
        split_data.append({"category": category, "reason": entry["reason"]})

# Create a DataFrame
df = pd.DataFrame(split_data)

# Count the number of complaints in each category
category_counts = df['category'].value_counts().sort_index()

# Plot a pie chart
plt.figure(figsize=(8, 8))
plt.pie(category_counts, labels=category_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Pie Chart: Distribution of Complaints by Category')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#complaints data
data = [
    {"category": "#webdev, #business", "reason": "This is a scam"},
    {"category": "#anime, #testing", "reason": "This user is fake"},
    {"category": "#testing", "reason": "Not trustworthy"},
    {"category": "#webdev, #business", "reason": "this is fake"},
    {"category": "#health, #gym", "reason": "this is fake"},
]

# Split entries with multiple categories into individual entries
split_data = []
for entry in data:
    categories = entry["category"].split(", ")
    for category in categories:
        split_data.append({"category": category, "reason": entry["reason"]})

# Create a DataFrame
df = pd.DataFrame(split_data)

# Count the number of complaints in each category
category_counts = df['category'].value_counts().sort_index()

# Plot a bar graph
plt.figure(figsize=(10, 6))
sns.barplot(x=category_counts.index, y=category_counts.values.astype(int), palette="viridis")
plt.title('Bar Graph: Number of Complaints in Each Category')
plt.xlabel('Category')
plt.ylabel('Number of Complaints')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# complaints data
data = [
    {"category": "#webdev, #business", "reason": "This is a scam"},
    {"category": "#anime, #testing", "reason": "This user is fake"},
    {"category": "#testing", "reason": "Not trustworthy"},
    {"category": "#webdev, #business", "reason": "this is fake"},
    {"category": "#health, #gym", "reason": "this is fake"},
]

# Split entries with multiple categories into individual entries
split_data = []
for entry in data:
    categories = entry["category"].split(", ")
    for category in categories:
        split_data.append({"category": category, "reason": entry["reason"]})

# Create a DataFrame
df = pd.DataFrame(split_data)

# Plot a scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='category', y='reason', data=df)
plt.title('Scatter Plot: Category vs Reason')
plt.xlabel('Category')
plt.ylabel('Reason')
plt.show()
```
